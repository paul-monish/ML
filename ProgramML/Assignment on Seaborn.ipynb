## Assignment on Seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1. Write a Python program to create a plot to get a general Statistics of Iris data. (Using the dataset “iris.csv”).

```python
df = sns.load_dataset("iris")
df
```

#### 2. Write a Python program to create a graph to find relationship between the sepal length and width. (Using the dataset “iris.csv”).

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=df)
plt.show()
```

#### 3. Write a Python program to create a join plot using "hexbin" to describe individual distributions on the same plot between Sepal length and Sepal width. (Using the dataset “iris.csv”).

```python
fig = sns.jointplot(x="sepal_length", y="sepal_width", kind="hex", color="blue", data=df)
plt.show()
```

#### 4. Write a Python program to create a joinplot and add regression and kernel density fits using "reg" to describe individual distributions on the same plot between Sepal length and Sepal width. (Using the dataset “iris.csv”).

```python
fig = sns.jointplot(x="sepal_length", y="sepal_width", kind="reg", color="blue", data=df)
plt.show()
```

#### 5. Write a Python program to find the correlation between variables of iris data. Also create a heatmap using Seaborn to present their relations (Using the dataset “iris.csv”).

```python
x = df.iloc[:,0:4]
f, ax = plt.subplots(figsize=(10,8))
corr = x.corr()
print(corr)
sns.heatmap(corr, mask=np.zeros_like(corr), 
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax, linewidths=.5)
plt.show()
```
