#### 4. Copy the datas from csv file dirtydata.csv and replace all the null calories value to 120 and also try to change the null date format to a valid date.


```python
df3=pd.read_csv("F:\BTech\ADV. PYTHON LAB\dataset\dirtydata.csv")
df31=df3.copy(deep=True)
df31['Calories'].fillna(120,inplace=True)
df31['Date']=pd.to_datetime(df31['Date'])
df31
```

#### 5. Take a data frame df2 and replace all Nan rows with the mode value for the column calories. Data set is dirtydata.

```python
df32=df3.copy(deep=True)
df32['Calories'].fillna(df32['Calories'].mode()[0],inplace=True)
df32
```

#### 6. Take df2 and for all the index values for the column calories if it is > 250 replace the calorie value with 250.

```python
for i in df32.index:
    if df32.loc[i,'Calories']>250:
        df32.loc[i,'Calories']=250
df32
```
